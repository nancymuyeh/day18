```python
import numpy as np
import matplotlib.pyplot as plt
def g(x):
    return (x**2) + 2*x + 5
plt.plot(x, g(x))
plt.title("Plot of  x^2 + 2x + 5")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 400)
def f(x):
    return np.sin(x**2 + 2*x + 5) * np.exp(-0.1*x)
plt.plot(x, f(x))
plt.title("Plot of f(x) = sin(x^2 + 2x + 5) x e^{-0.1x}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
